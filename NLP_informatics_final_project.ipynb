# **Programming project**

This programming project is an interactive verb conjugator. The database already contains four entries for the German language. New entries can be added or existing ones corrected, and the database can be displayed as a whole or partly.
Run all cells to prepare the verb conjugator. The last cell will start the verb conjugator.



```python
# initial database

pronouns = ["ich", "du", "er/sie/es", "wir", "ihr", "sie/Sie"]
regular_affixes = ["e","st","t","en","t","en","ge-t"]
regular_affixes_2 = ["e","st","t","n","t","n"]
persons_numbers = {"1sg":0,"2sg":1,"3sg":2,"1pl":3,"2pl":4,"3pl":5}
database = {
    "sein":{
        "present":["bin","bist","ist","sind","seid","sind"],
        "past-participle":["gewesen","sein"]},
    "haben":{
        "present":["habe","hast","hat","haben","habt","haben"],
        "past-participle":["gehabt","haben"]},
    "kochen":{
        "present":["koche","kochst","kocht","kochen","kocht","kochen"],
        "past-participle":["gekocht","haben"]},
    "schwimmen":{
        "present":["schwimme","schwimmst","schwimmt","schwimmen","schwimmt","schwimmen"],
        "past-participle":["geschwommen","sein"]},
    }
```

```python
# all messages

welcome_message = """
    Hello. I am an interactive verb conjugator.
    You can ask for a specific conjugation in my database.
    If the verb does not exist, you have the option to add a new entry.
    You can also correct an existing entry.
    To know about all your options, please type "help".
"""

goodbye_message = """
    It was a pleasure working with you! See you next time!
"""

command_unknown = """
    I am sorry, I do not understand this command.
    Please use one of the options available.
    To show all options, type "help".
"""

options = """
    help : shows all options

    conjugate [verb] [tense] [person/number] : prints the requested conjugation.
        Possible tenses are "present" and "past-participle".
        e.g. "conjugate haben present 1sg" will print "ich habe"

    display [optional:verb] [optional:tense] : prints the requested subpart of the database.
        e.g. "display" will print the whole database,
        "display haben" will print the entry for "haben",
        "display haben present" will print the present tense paradigm for "haben".

    add [verb] : guides you through adding a new verb into the database.
        e.g. "add schreiben"

    correct [verb] : guides you through correcting an existing entry.
        e.g. "correct haben"

    exit : ends program.
"""

suggest_adding_new_entry = """
    I am sorry, this verb does not exist in my database.
    If you want to see which verbs are part of my database, type "display".
    If you want to add an entry, type "add [verb]".
"""

trying_to_access_nonexisting_data = """
    I am sorry, this is not part of my database.
    Type "display" to see the whole database and identify accessible subparts.
"""

trying_to_add_existing_verb = """
    This verb already exists, so I will not add a new entry.
    If you want to see the existing entry, type "display [verb]".
    If you want to correct it, type "correct [verb]".
"""

request_auxiliary_for_new_entry = """
    What is the auxiliary for the past participle of this verb? Type "haben" or "sein".
"""

ask_if_verb_is_regular = """
    Is it a regular verb in the present tense and the past participle?
    Type "yes" or "no".
"""

request_irregular_forms = """
    Please indicate the verb paradigm for the present tense and the past participle.
    Separate each conjugation with a whitespace.
    e.g. "bin bist ist sind seid sind gewesen"
    If some of the conjugations are regular, you can replace them by a dash ("-").
    e.g. "- hast hat - - - -"
"""

request_interpretable_answer = """
    Please answer according to my previous instructions.
"""

new_entry_added = """
    I have added the new entry. Make sure everything is correct.
    If there is an error in the entry, you can type "correct [verb]" to correct it.
    The new entry looks like this:
"""

ask_for_part_to_be_corrected = """
    Which part of the entry do you want to correct?
    Type "present" or "past-participle".
    The current entry is:
"""

request_present_forms_for_correction = """
    Please provide the correct verb paradigm for the present tense.
    Separate each conjugation with a whitespace.
    For those conjugations that are already correct you can type dash ("-").
    e.g.
    "bin bist ist - - -"
    If you do not want to change anything, type "- - - - - -"
    The current entry is:
"""

request_auxiliary_and_participle_for_correction = """
    Please provide the correct participle and auxiliary for the present participle.
    Separate them with a whitespace.
    If one of it is already correct, you can replace it by a dash ("-").
    e.g.
    "gewesen sein", or "gewesen -"
    If you do not want to change anything, type "- -"
    The current entry is:
"""

entry_corrected = """
    I have modified the entry according to your input. Make sure, everything is correct now.
    If there is still an error in the entry, you can type "correct [verb]" to correct it.
    The entry looks like this now:
"""
```

```python
# built-in functions used for the assignment
# (only for better readability)

import pprint
pp = pprint.PrettyPrinter(indent=4)
```

```python
# auxiliary functions

def show_options():
    print(options)

def conjugate(parameters):
    verb, tense, subject = parameters
    if (verb not in database) or (tense not in database[verb]) or (subject not in persons_numbers):
        print(suggest_adding_new_entry)
        return
    person_number = persons_numbers[subject]
    pronoun = pronouns[person_number]
    if tense == "present":
        conjugation = database[verb][tense][person_number]
    else: # tense == "past-participle"
        participle = database[verb][tense][0]
        auxiliary = database[verb][tense][1]
        auxiliary_conjugation = database[auxiliary]["present"][person_number]
        conjugation = f"{auxiliary_conjugation} {participle}"
    print("   ",pronoun,conjugation)

def display_subpart(dictionary_path):
    current_subpart = database
    for level in dictionary_path:
        if level in current_subpart:
            current_subpart = current_subpart[level]
        else:
            print(trying_to_access_nonexisting_data)
            return
    print("", end ="    ") # to indent the answer without adding a new line
    pp.pprint(current_subpart)

def add_entry(verb,auxiliary,irregular_forms):
    database[verb] = {}
    verb_group_2 = False
    if verb[-2] != "e" and verb[-3] == "e": # some German verbs end in -eln, -ern
        stem = verb[:-1]
        verb_group_2 = True
    else:
        stem = verb[:-2]
    conjugations = ["","","","","",""]
    for i in range(6):
        if irregular_forms[i] == "-":
            if verb_group_2:
                conjugations[i] = stem+regular_affixes_2[i]
            else:
                conjugations[i] = stem+regular_affixes[i]
        else:
            conjugations[i] = irregular_forms[i]
    database[verb]["present"] = conjugations
    if irregular_forms[6] == "-":
        circumfix = regular_affixes[6].split("-")
        participle = circumfix[0] + stem + circumfix[1]
    else:
        participle = irregular_forms[6]
    database[verb]["past-participle"] = [participle,auxiliary]

def correct_entry(verb,correction):
    if len(correction) == 6:
        tense = "present"
    else:
        tense = "past-participle"
    for i in range(len(correction)):
        if correction[i] != "-":
            database[verb][tense][i] = correction[i]
```

```python
# central function

def start_interactive_conjugator():
    command = input(welcome_message)

    while command != "exit":

        if command == "help":
            show_options()

        elif command.split()[0] == "conjugate" and len(command.split()) == 4:
            conjugate(command.split()[1:])

        elif command.split()[0] == "display" and len(command.split()) <= 3:
            display_subpart(command.split()[1:])

        elif command.split()[0] == "add" and len(command.split()) == 2:
            verb = command.split()[1]
            if verb in database:
                print(trying_to_add_existing_verb)
            else:
                auxiliary = input(request_auxiliary_for_new_entry)
                while auxiliary not in ["haben","sein"]:
                    auxiliary = input(request_interpretable_answer)
                is_regular = input(ask_if_verb_is_regular)
                while is_regular not in ["yes","no"]:
                    is_regular = input(request_interpretable_answer)
                if is_regular == "yes":
                    irregular_forms = ["-","-","-","-","-","-","-"]
                else:
                    irregular_forms = input(request_irregular_forms).split()
                    while len(irregular_forms) != 7:
                        irregular_forms = input(request_interpretable_answer).split()
                add_entry(verb,auxiliary,irregular_forms)
                print(new_entry_added)
                display_subpart([verb])

        elif command.split()[0] == "correct" and len(command.split()) == 2:
            verb = command.split()[1]
            if verb not in database:
                print(suggest_adding_new_entry)
            else:
                print(ask_for_part_to_be_corrected)
                display_subpart([verb])
                part_to_be_corrected = input()
                while part_to_be_corrected not in ["present","past-participle"]:
                    part_to_be_corrected = input(request_interpretable_answer)
                if part_to_be_corrected == "present":
                    print(request_present_forms_for_correction)
                    display_subpart([verb,"present"])
                    correction = input().split()
                    while len(correction) != 6:
                        print(request_interpretable_answer)
                        correction = input().split()
                if part_to_be_corrected == "past-participle":
                    print(request_auxiliary_and_participle_for_correction)
                    display_subpart([verb,"past-participle"])
                    correction = input().split()
                    while len(correction) != 2:
                        print(request_interpretable_answer)
                        correction = input().split()
                correct_entry(verb,correction)
                print(entry_corrected)
                display_subpart([verb])

        else:
            print(command_unknown)

        command = input()

    print(goodbye_message)
```

```python
start_interactive_conjugator()
```
